```python
def collision_checker(ego_vehicle: State, obstacles: List[State]) -> List[bool]:
    collisions = [False] * len(obstacles)
    n_circles = len(CIRCLE_OFFSETS)

    for obs, actor in enumerate(obstacles):
        collision = False
        for c in range(n_circles):
            # Circle placement
            ego_circle_center = Location()
            ego_circle_center.x = ego_vehicle.location.x + CIRCLE_OFFSETS[c] * math.cos(ego_vehicle.rotation.yaw)
            ego_circle_center.y = ego_vehicle.location.y + CIRCLE_OFFSETS[c] * math.sin(ego_vehicle.rotation.yaw)
            for c2 in range(n_circles):
                actor_circle_center = Location()
                actor_circle_center.x = actor.location.x + CIRCLE_OFFSETS[c2] * math.cos(actor.rotation.yaw)
                actor_circle_center.y = actor.location.y + CIRCLE_OFFSETS[c2] * math.sin(actor.rotation.yaw)

                # Distance from circles to obstacles/actor
                dist = math.sqrt((ego_circle_center.x - actor_circle_center.x) ** 2
                                 + (ego_circle_center.y - actor_circle_center.y) ** 2)

                # Consider a collision when the distance between the 2 centers is smaller than the sum of their radii
                collision = (dist < (CIRCLE_RADII[c] + CIRCLE_RADII[c2]))
                if collision:
                    break
        collisions[obs] = collision

    return collisions
```
