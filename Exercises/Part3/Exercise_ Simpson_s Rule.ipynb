```python
def calc_theta(s: float, a: Tuple[float, float, float, float]) -> float:
    """
    Calculates theta(s) for a given s and coefficients a.
    theta(s) = a3*s^4/4 + a2*s^3/3 + a1*s^2/2 + a0*s
    """
    a0, a1, a2, a3 = a
    return (((a3  *s / 4 + a2 / 3)*  s + a1 / 2)  *s + a0)*  s;
```

```python
def integrate_by_simpson(func: List[float], dx: float, n: int) -> float:
    """
    Uses Simpson's Rule to numerically integrate the given function.
    func - List of function values.
    dx - Step size.
    n - Number of intervals.
    """
    if n == 0:
        return 0.0
    
    sum_even = 0.0
    sum_odd = 0.0
    for i in range(1,n):
        if i % 2 != 0:
            sum_odd += func[i]
        else:
            sum_even += func[i]
    
    result = (dx / 3) * (func[0] + 4 * sum_odd + 2 * sum_even + func[n])
    return result
```

```python
def generate_f_s0_sn(delta_s: float, a: Tuple[float, float, float, float], n: int) -> List[float]:
    """
    Generates a list of function values to be integrated using Simpson's Rule.
    delta_s - Step size.
    a - Coefficients.
    n - Number of intervals.
    """
    f_s = []
    for i in range(n+1):
        s_i = i * delta_s
        theta_i = calc_theta(s_i, a)
        f_s.append(math.cos(theta_i))
    return f_s
```
